```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
"""use numpy to generate random data"""
import numpy as np
x=np.linspace(0,10,25)
y=x*x+2
print(x)
print(y)
print(np.array([x,y]).reshape(25,2).reshape(2,25))
```

```python
"""it only takes 1 command to draw """
pylab.plot(x,y,'r') # 'r' stands for red
```

```python
""" drawing a subgraph"""
pylab.subplot(1,2,1) # the contents of the brackets represent (rows,columns,indexes)
pylab.plot(x,y,'r--')# the third parameter here determines color and line style
pylab.subplot(1,2,2)
pylab.plot(y,x,'g+-')
```

```python
#(1,2,1)and (1,2,2) are significant if we want tiled subplots,without which both the graphs will in the smae plot
# for example
pylab.subplot(1,2,1) # the contents of the brackets represent (rows,columns,indexes)
pylab.plot(x,y,'r--')# the third parameter here determines color and line style
pylab.subplot(1,2,1)
pylab.plot(y,x,'g+-')
```

```python
from matplotlib import pyplot as plt
```

```python
fig=plt.figure()
axis=fig.add_axes([0.5,0.1,0.8,0.8]) # control the left,right,width,height of the canvas (from 0 to 1)
axis.plot(x,y,'r')
```

```python
# subgraphs
fig,axes=plt.subplots(nrows=1, ncols=2) # submap if of 1 row, 2 columns
for ax in axes:
    ax.plot(x,y,'r')
```

```python
# draw a picture, or graph, inside another graph
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8]) #big axes
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) # small canvas    # left,bottom,width,height
axes1.plot(x,y,'r')
axes2.plot(x,y,'b')
```

```python
fig=plt.figure(figsize=(10,9),dpi=100) # new graphic object # dpi (dasphiraze)
fig.add_subplot()
plt.plot(x,y,'r')
```

```python
fig,axes=plt.subplots()
axes.set_title("Title")
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.plot(x,y,'r')
```

### Other 2D Graphics

```python
n=np.array([0,1,2,3,4,5])
fig,axes=plt.subplots(1,4,figsize=(16,5))
axes[0].set_title('scatter plot')
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))
axes[1].set_title('step plot')
axes[1].step(n,n**2,lw=2)
axes[2].set_title('bar plot')
axes[2].bar(n,n**2,align="center",width=0.5,alpha=0.5)
axes[3].set_title('fill_between plot')
axes[3].fill_between(x,x**2,x**3,color="green",alpha=0.5)
```

```python
"""draw a radar chart"""
fig=plt.figure(figsize=(6,6))
ax=fig.add_axes([0.0,0.0,0.6,0.6],polar=True)
t=np.linspace(0,2*np.pi,100)
ax.plot(t,t,color='blue',lw=3)
```

```python
"""draw a histogram"""
n=np.random.randn(1000)
fig,axes=plt.subplots(1,2,figsize=(10,4))
axes[0].set_title("Default histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative detailed histogram")
axes[1].hist(n,cumulative=True, bins=50)
```

```python
# draw contour image
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta=0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
X,Y=np.meshgrid(x,y)
Z1=np.exp(-X**2 -Y**2)
Z2=np.exp(-(X-1)**2 -(Y-1)**2)
Z=(Z1-Z2)*2
print(X)
print(Y)
```

```python
fig,ax=plt.subplots()
CS=ax.contour(X,Y,Z)
ax.clabel(CS,inline=1,fontsize=10)
ax.set_title('contour map')
```
